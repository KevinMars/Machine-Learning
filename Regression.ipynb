### Challenge 1

Fit a model to your data. Write a python function to simulate outcomes.  
(The sigma of the normal noise distribution will be the std deviation of the residuals.)  
For the same observed input variables, simulate the outcome.  
Plot the observed data and the simulated data.

```python
import statsmodels.formula.api as sm
from sklearn.metrics import mean_squared_error

# Fit a model
X = movies[['Budget', 'Ones']]
Y = movies['DomesticTotalGross']
model = sm.OLS(Y, X).fit()

# Read the Beta_1 and Beta_0 coefficients
print 'Coefficients of the model:\n', model.params
print
beta_1, beta_0 = model.params

# To simulate the normally distributed error epsilon,
# we need to know the mu (mean) and sigma (stdev)
# of that normal distribution. The mean is zero,
# and the stdev is the stdev of the residuals
# (differences between prediction and observed values)
# This is also the same thing as the square root of the
# "mean squared error".
residual_stdev = np.std(model.resid)
print 'stdev of residuals:', residual_stdev

# Just to demonstrate that this is exactly the same thing as the sqrt(MSE):
print 'square root of MSE:', np.sqrt(mean_squared_error(Y, model.predict(X)))

# Make a function to simulate the model
# Y = beta_0 + beta_1 * budget
import random
def simulate_gross(budget_array):
    epsilon_array = [random.gauss(0, residual_stdev) for x in budget_array]
    sim_gross_array = beta_0 + beta_1 * budget_array + epsilon_array
    return sim_gross_array

# The observed and simulated data to plot
X = movies['Budget']
observed_Y = movies['DomesticTotalGross']
simulated_Y = simulate_gross(movies['Budget'])
model_prediction = model.predict(movies[['Budget','Ones']])

# Plot!
obs = plt.scatter(X, observed_Y, color='blue')
sim = plt.scatter(X, simulated_Y, color='green')
mod = plt.scatter(X, model_prediction, color='red', marker='s')

# Add a legend
plt.legend((mod,obs,sim,mod),
           ('Prediction', 'Observed', 'Simulated'),
           loc='upper left')
```

### Challenge 2

Generate (fake) data that is linearly related to log(x).   
Basically write an underlying model just like in challenge 1, but instead of a fitted model, you are making this model up. It is of the form B0 + B1log(x) + epsilon. You are making up the parameters.  
Simulate some data from this model. Then fit two models to it:  
a) quadratic [second degree polynomial]   
b) logarithmic [log(x)]  
(the second one should fit really well, since it has the same form as the underlying model!)  

```python
# A function to simulate Y values for given X values
# It simulates them according to the model
# beta_0 + beta_1 log(X) + epsilon
# The beta coefficients and the size of noise is up to you
def simulate_logx_model(x_array, beta_0, beta_1, noise_std):
    epsilon_array = [random.gauss(0, noise_std) for x in x_array]
    y_array = beta_0 + beta_1 * np.log(x_array) + epsilon_array
    return y_array


# Create fake data (200 points)
# I choose the beta 0, beta 1 and noise std values as
# 500, 5 and 3, you can play around with these
# It is your fake data :)
X = np.linspace(10,1e4, num = 200)
Y = simulate_logx_model(X, beta_0=500, beta_1=5, noise_std=3)

# Start a figure with 1x3 subplots
plt.figure(figsize=(15,5))

# Plot fake data
plt.scatter(X,Y)

# Fit and plot quadratic model
# ---first create an input matrix with X^2, X and Ones features
Xsq = X**2
Ones = np.ones(200)
X_input = np.column_stack((Xsq,X,Ones))
# ---now fit
quad_model = sm.OLS(Y,X_input).fit()
#---plot
plt.subplot(121)
plt.scatter(X,Y)
plt.plot(X,quad_model.predict(X_input), 'r-', linewidth=5)
plt.title("Quadratic Model")

# Fit and plot log x model
# ---input matrix with logX and Ones features
logX = np.log(X)
Ones = np.ones(200)
X_input = np.column_stack((logX,Ones))
# ---fit
logx_model = sm.OLS(Y,X_input).fit()
#---plot
plt.subplot(122)
plt.scatter(X,Y)
plt.plot(X,logx_model.predict(X_input), 'g-', linewidth=5)
plt.title("Logarithmic Model")
```

#### Challenge 4

For one continuous feature (like budget, choose one that strongly affects the outcome), try polynomial fits from 0th (just constant) to 7th order (highest term x^7). Over the x axis of model degree (8 points), plot:  
a) training error  
b) test error  
c) R squared  
d) AIC  


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Make a function to create an input matrix with the
# polynomial degrees of X
# For example, polynomial_features(X, 3) returns a matrix
# with the vectors X, X**2, X**3
def polynomial_features(x_array,degree):
    if degree == 0:
        return np.ones(len(x_array)).reshape(-1,1)
    feature_vectors = []
    for power in range(1,degree+1):
        feature_vectors.append( x_array**power )
    feature_matrix = np.column_stack( feature_vectors )
    return feature_matrix

# Another function to fit a polynomial model for a 
# given X
def fit_poly_model(X, Y, degree):
    poly_input = polynomial_features(X,degree)
    poly_model = LinearRegression()
    poly_model.fit(poly_input, Y)
    return poly_model

#For each degree from 1 to 7, fit a polynomial and record the test and training errors
degrees = []
training_errors = []
test_errors = []
r2s = []
aics=[]

for degree in range(0,8):

    # fit the model
    X = train_set['Budget']
    Y = train_set['DomesticTotalGross']
    poly_model = fit_poly_model(X, Y, degree)

    # training error
    X_train_input = polynomial_features(train_set['Budget'], degree)
    Y_train_pred = poly_model.predict( X_train_input )
    Y_train_true = train_set['DomesticTotalGross']
    train_err = mean_squared_error(Y_train_true, Y_train_pred)

    #test error
    X_test_input = polynomial_features(test_set['Budget'], degree)
    Y_test_pred = poly_model.predict( X_test_input )
    Y_test_true = test_set['DomesticTotalGross']
    test_err = mean_squared_error(Y_test_true, Y_test_pred)
    
    #R squared
    r_squared = r2_score(Y_train_true, Y_train_pred)
    r2s.append(r_squared)
    
    #AIC
    # --- add 'ones' to input for statsmodels
    statsmodels_input = np.column_stack((X_train_input,np.ones(len(X)).reshape(-1,1)))
    # --- fit statsmodels to get aic
    aic = sm.OLS(Y_train_true, statsmodels_input).fit().aic
    # --- save it
    aics.append(aic)
    
    print '[Degree %i]   Training err: %g    Test err: %g' % (degree, train_err, test_err)
    degrees.append(degree)
    training_errors.append(train_err)
    test_errors.append(test_err)
    
    
# Plot Errors
plt.figure(figsize=(9,15))
plt.subplot(311)
tr, = plt.plot(degrees, training_errors, color='red')
te, = plt.plot(degrees, test_errors, color='green')
plt.legend((tr,te), ('Training error', 'Test Error'), loc='upper right')
plt.title("Errors")
plt.xlabel("model degree")

# Plot R^2
plt.subplot(312)
plt.plot(degrees, r2s, 'b-')
plt.title("R squared")
plt.xlabel("model degree")

# Plot AIC
plt.subplot(313)
plt.plot(degrees, aics, 'm-')
plt.title("AIC")
plt.xlabel("model degree")
```

### Challenge 5

Fit a model to only the first 5 of your data points (m=5). Then to first 10 (m=10). Then to first 15 (m=15). In this manner, keep fitting until you fit your entire training set. For each step, calculate the training error and the test error. Plot both (in the same plot) over m. This is called a learning curve.

```python
random.seed(4444)

# Function to fit a linear model to only to m randomly selected points
def fit_model_to_m_points(dataframe, m, input_cols, outcome_col):
    random_indices = list(dataframe.index)
    random.shuffle(random_indices)
    # only use m (randomly selected) rows
    fitting_set = dataframe.ix[ random_indices[:m] ]
    Y = fitting_set[outcome_col]
    X = fitting_set[input_cols]
    return X, Y, sm.OLS(Y,X).fit()


# For each m value (number of training set points), fit a model and compute errors
def make_learning_curve_up_to(m, train_set, test_set):
    ms = []
    train_errs = []
    test_errs = []
    for m in range(4,m):
        ms.append(m)
        X_train, Y_train, model = fit_model_to_m_points(train_set, m, ['Budget','Ones'], 'DomesticTotalGross')
    
        # training error
        Y_train_pred = model.predict( X_train )
        Y_train_true = Y_train
        train_err = mean_squared_error(Y_train_true, Y_train_pred)
        train_errs.append( train_err )
    
        #test error
        Y_test_pred = model.predict( test_set[['Budget','Ones']] )
        Y_test_true = test_set['DomesticTotalGross']
        test_err = mean_squared_error(Y_test_true, Y_test_pred)
        test_errs.append( test_err )
    
    #done
    return ms, train_errs, test_errs
    
# Plot full learning curve (~500 points)
draw_up_to = len(train_set)
ms, train_errs, test_errs = make_learning_curve_up_to(draw_up_to, train_set, test_set)

plt.figure(figsize=(18,5))
plt.subplot(121)
tr, = plt.plot(ms, train_errs, color='red')
te, = plt.plot(ms, test_errs, color='green')
plt.legend((tr,te), ('Training error', 'Test Error'), loc='upper right')
plt.title("Learning curve")
plt.xlabel("Number of data points")


# Plot only the first 75 points (to zoom in on the conversion)
draw_up_to = 75
ms, train_errs, test_errs = make_learning_curve_up_to(draw_up_to, train_set, test_set)

plt.subplot(122)
tr, = plt.plot(ms, train_errs, color='red')
te, = plt.plot(ms, test_errs, color='green')
plt.legend((tr,te), ('Training error', 'Test Error'), loc='upper right')
plt.title("Learning curve")
plt.xlabel("Number of data points")

```
