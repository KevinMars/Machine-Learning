# DecisionTree Challenges

```python
import pandas as pd
import numpy as np

df = pd.io.parsers.read_csv('house-votes-84.csv', header=None)
```

```python
df = df.apply(lambda x: x.replace('y',1).replace('n',0)).replace('?', np.nan).replace('democrat',1).replace('republican',0)
```

```python
for x in range(1,17):
    colmean = df[x].mean()
    df.loc[np.isnan(df[x]), x] = colmean
df.head()
```

```python
from sklearn.cross_validation import train_test_split
X = df[range(1,17)]
Y = df[0]
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.30, random_state=4444)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import BaggingClassifier

decision_tree = DecisionTreeClassifier().fit(X_train,Y_train)
Y_pred_tree = decision_tree.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred_tree)
print accuracy
```

### Challenge 1

```python
from sklearn.externals.six import StringIO  
import pydot 
from sklearn.tree import export_graphviz

dot_data = StringIO() 
export_graphviz(decision_tree, out_file=dot_data) 
graph = pydot.graph_from_dot_data(dot_data.getvalue()) 
graph.write_pdf('decision_tree.pdf')
```

```python
titanic_train = pd.io.parsers.read_csv('train.csv')
titanic_train = titanic_train.apply(lambda x: x.replace('male',0).replace('female',1))
titanic_train['Embarked'] = titanic_train['Embarked'].map({'S': 2, 'C': 1, 'Q':0})
titanic_train.head()
```

## Titanic Accuracy

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.cross_validation import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

accuracy_log = np.mean(cross_val_score(LogisticRegression(), tX,tY, cv=10))
print accuracy_log

accuracy_tree = np.mean(cross_val_score(DecisionTreeClassifier(), tX,tY, cv=10))
print accuracy_tree

accuracy_forest = np.mean(cross_val_score(RandomForestClassifier(), tX,tY, cv=10))
print accuracy_forest

accuracy_svc = np.mean(cross_val_score(SVC(), tX,tY, cv=10))
print accuracy_svc

accuracy_nb = np.mean(cross_val_score(GaussianNB(), tX,tY, cv=10))
print accuracy_nb
```
